```python
import os
import glob
import rasterio as rio


png = 'D:\\python_data\\img\\Kuppam.png'
tif = 'D:\\python_data\\img\\Kuppam_15_12_rf.tif'
img = rio.open(png)
img= img.read([1,2,3])

with rio.open('D:\\python_data\\img\\Kuppam_15_12_rf.tif') as nip:
    
    with rio.open('D:\\python_data\\img\\Kuppam_georeferenci.tif',
                  'w',
                  driver ='Gtiff',
                  count = img.shape[0],
                  height = img.shape[1],
                  width = img.shape[2],
                  dtype = img.dtype,
                  crs = nip.crs,
                  transform = nip.transform
                 )as df :
        df.write(img)
    
    
```
